```python
cars_df['Total Speed'] = cars_df['Total Speed'].str.extract(r'(\d+\.?\d*)')[0]
cars_df['Total Speed'] = pd.to_numeric(cars_df['Total Speed'], errors='coerce')
```

```python
cars_df['Total Speed'].plot.hist()
```

```python
agg = cars_df.groupby('Seats')['Total Speed'].mean().reset_index()
```

```python
heatmap_data = agg.pivot_table(index='Seats', values='Total Speed')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 8))
sns.heatmap(heatmap_data, annot=True, cmap='YlOrBr')
plt.title('Average Total Speed by Number of Seats')
plt.xlabel('Average Speed')
plt.ylabel('Number of Seats')
plt.show()
```

```python
# Group and compute mean
grouped = cars_df.groupby('Cleaned_Seat')['Total Speed'].mean().sort_index()

# Convert to DataFrame for heatmap
heatmap_data = pd.DataFrame(grouped)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 10))
sns.heatmap(heatmap_data, annot=True, cmap='YlOrBr', cbar_kws={'label': 'Average Speed'})
plt.title("Average Total Speed by Normalized Number of Seats")
plt.ylabel("Number of Seats")
plt.xlabel("Total Speed")
plt.show()
```
